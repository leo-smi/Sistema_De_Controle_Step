```
lw = 3 # expessura da linha
################################################################################
xlim1 = 3 # range do tempo para o step1
plt.subplot(211) # primeira imagem do subplot
G1 = tf([1], [2, 3]) # função de transferencia 
G1fb = feedback(G1) # realimentação da função de transfêrencia
(y1, t1) = step(G1fb, T=np.linspace(0, xlim1, 50)) #T é um vetor de tempo 
lgd1 = "Gfb1(s)" # legenda da curva da figura 1
plt.plot(t1, y1, label=lgd1, linewidth=lw, color="red") # plota a figura 1 (t x y)
plt.grid() # plota a grade na figura 1
plt.ylabel('Amplitude') # texto do eixo y
plt.xlabel('Tempo [s]') # texto do eixo x
plt.title("Resposta ao degrau (primeira ordem)") # título da imagem
plt.legend() # mostra a legenda pertencente a figura 1
plt.xlim(0, xlim1) # seta o range do eixo x
plt.ylim(0, 0.3) # seta o range do eixo y
################################################################################
xlim2 = 8
plt.subplot(212)
G2 = tf([1], [2, 3, 5])
G2fb = feedback(G2)
(y2, t2) = step(G2fb, T=np.linspace(0, xlim2, 50))
plt.plot(t2, y2, label="Gfb2(s)", linewidth=lw)
plt.grid()
plt.ylabel('Amplitude')
plt.xlabel('Tempo [s]')
plt.title("Resposta ao degrau (segunda ordem)")
plt.legend()
plt.xlim(0, xlim2)
plt.ylim(0, 0.3)
################################################################################
plt.tight_layout() # ajusta o espaço entre os gráficos
################################################################################
plt.savefig('/content/gdrive/My Drive/Colab Notebooks/controle01.jpeg', dpi=300)
```
